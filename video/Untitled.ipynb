```python
import pandas as pd

# Excel dosyasını oku
df = pd.read_excel("orta2.xlsx")  # Excel dosyasının adı

# HTML kodlarını oluştur
html_codes = []

for index, row in df.iterrows():
    title = row['Title']
    video_url = row['Video url']
    
    # YouTube video ID'yi almak
    video_id = video_url.split('v=')[-1]
    
    html_code = f'''
    <div class="section"> 
        <iframe src="https://www.youtube.com/embed/{video_id}" allowfullscreen></iframe>
        <a href="{video_url}" target="_blank">{title}</a>
    </div>
    '''
    html_codes.append(html_code)

# Oluşturulan HTML kodlarını dosyaya yaz
with open("output.html", "w", encoding="utf-8") as file:
    file.write("\n".join(html_codes))

print("HTML kodları başarıyla oluşturuldu!")
```
